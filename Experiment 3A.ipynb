```r
library(ggplot2)
```

```r
# Load the dataset
car_data <- read.csv("Datasets/Car_prices.csv")

# Keep only the required columns
car_data <- car_data[, c("carwidth", "cylindernumber", "enginesize", "boreratio", "horsepower", "highwaympg", "price")]

# Check the structure of the dataset
str(car_data)
```

```r
unique_values <- unique(car_data$cylindernumber)
print(unique_values)
```

```r
# Create a mapping for string values to numeric values
mapping <- c("four" = 4, "six" = 6, "five" = 5, "three" = 3, "twelve" = 12, "two" = 2, "eight" = 8)

# Replace string values with numeric values
car_data$cylindernumber <- mapping[as.character(car_data$cylindernumber)]

str(car_data)
```

```r
# Summary statistics
summary(car_data)
```

```r
# Visualize distributions
hist(car_data$carwidth, main="Car Width")
hist(car_data$cylindernumber, main="Cylinder Number")
hist(car_data$enginesize, main="Engine Size")
hist(car_data$boreratio, main="Bore Ratio")
hist(car_data$horsepower, main="Horsepower")
hist(car_data$highwaympg, main="Highway MPG")
```

```r
# Check correlations
cor(car_data)
```

```r
# Scatterplot matrix
pairs(car_data)
```
