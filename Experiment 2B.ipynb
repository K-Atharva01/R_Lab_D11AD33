Dataset used - https://www.kaggle.com/datasets/sashankpillai/spotify-top-200-charts-20202021

```r
spotify_data <- read.csv("Datasets/spotify_dataset.csv")

# str(spotify_data)

spotify_data$Streams <- gsub("[^0-9]", "", spotify_data$Streams)

spotify_data$Streams <- as.numeric(as.character(spotify_data$Streams))

song_streams <- aggregate(Streams ~ `Song.Name`, data = spotify_data, FUN = sum)

top_songs <- head(song_streams[order(-song_streams$Streams), ], 5)
```

```r
# Preprocess the genre data to remove brackets and quotation marks
cleaned_genres <- gsub("\\[|\\]|'", "", spotify_data$Genre)
all_genres <- unlist(strsplit(cleaned_genres, ", "))  

# Count the frequency of each genre
genre_counts <- table(all_genres)

# Select the top N most frequent genres (e.g., top 10)
top_genres <- head(sort(genre_counts, decreasing = TRUE), 10)

# Calculate percentages manually
genre_percentages <- top_genres / sum(top_genres) * 100

# Format percentages as character strings
percentage_labels <- paste0(round(genre_percentages, 1), "%")

# Create labels with percentages
labels_with_percentages <- paste(names(top_genres), " (", percentage_labels, ")", sep = "")

# Create the pie chart with labels including percentages
pie(top_genres, 
    labels = labels_with_percentages,
    main = "Top 10 Genres in Top Songs",
    col = rainbow(length(top_genres)),
    cex.main = 2,
    cex.lab = 1.2)

```
