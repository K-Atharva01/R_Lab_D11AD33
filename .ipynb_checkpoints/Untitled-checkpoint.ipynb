```r
library(ggplot2)  
```

```r
dataset <- read.csv("Car_prices.csv")
str(dataset)
```

```r
head(dataset)
summary(dataset)
colSums(is.na(dataset))
```

## BarPlot

```r
library(ggplot2)

# Bar plot of CarName
ggplot(dataset, aes(x = CarName)) +
  geom_bar(fill = "skyblue", color = "black") +
  labs(title = "Distribution of Car Names", x = "Car Name", y = "Frequency")
```

## BoxPlot

```r
library(ggplot2)

# Boxplot of Price grouped by CarName
ggplot(dataset, aes(x = CarName, y = price, fill= price)) +
  geom_boxplot(fill = "skyblue", color = "black") +
  labs(title = "Distribution of Price by Car Name", x = "Car Name", y = "Price")
```

## ScatterPlot

```r
library(ggplot2)

# Scatterplot of LungCap vs Age colored by Smoke status
ggplot(dataset, aes(x = CarName, y = enginetype, color = price)) +
  geom_point() +
  labs(title = "Lung Capacity vs Age (Colored by Smoking Status)", x = "Age", y = "Lung Capacity")
```

## PairPlot

```r
# Load necessary libraries
library(ggplot2)
library(dplyr)

# Selecting numerical variables for the pair plot
numeric_data <- select(dataset, horsepower, citympg, peakrpm)

# Pair plot
pairs(numeric_data)
```

## Correlation Matrix

```r
# Calculate correlation coefficient matrix
correlation_matrix <- cor(select(dataset, horsepower, citympg, peakrpm))

# Print correlation coefficient matrix
print(correlation_matrix)
```

## Heatmap

```r
# Visualize correlation coefficients using heatmap

heatmap(correlation_matrix,
        main = "Correlation Heatmap of Numerical Variables",
        xlab = "Variables",
        ylab = "Variables",
        cex.axis = 1,  # Increase font size of axis labels
        margins = c(10, 10))
```
